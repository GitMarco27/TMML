## Converting a Keras model to a TF Lite Model
---
Marco Sanguineti, 2022

```python
import tensorflow as tf
tf.__version__
```

```python
input_lay = tf.keras.layers.Input(shape=[1])

lay_1 = tf.keras.layers.Dense(units=16, activation='selu')(input_lay)

out_lay = tf.keras.layers.Dense(units=1, activation='linear')(lay_1)
```

```python
model = tf.keras.models.Model(inputs= input_lay, outputs=out_lay)
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

model.compile(optimizer=optimizer, loss=tf.keras.losses.mean_squared_error)
```

```python
model.fit(x=[-1, 0, 1], y=[2, 3, 4], epochs=10)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
```

```python
tflite_model = converter.convert()
```

```python
tflite_model
```

```python
with open('model.tflite', 'wb') as model_:
  model_.write(tflite_model)
```
